# Gráfico detallado

En este último ejemplo de gráficas con Python voy a hacer un
gráfico y lo voy a documentar lo más que pueda para una futura
referencia. Tampoco exageradamente.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) # Tomamos el último dígito como true para agragarlo

S, C = np.sin(x), np.cos(x) # Asignamos dos variables separadas por comas en una misma línea
```

```python
plt.figure(figsize=(9,4.5))
plt.plot(x, S, 'b-', linewidth = 1.2, label = r'$\sin \; x$')
plt.plot(x, C, 'g-', linewidth = 1.2, label = r'$\cos \; x$')
plt.xlim(x.min()*1.10, x.max()*1.10) # Expandimos 10% el eje x
plt.ylim(S.min()*1.10, S.max()*1.10)

# Las marcas en los ejes se llaman 'ticks' ya sean x o y
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$',
           r'$\frac{\pi}{2}$', r'$\pi$'])

plt.yticks(np.linspace(-1,1,3, endpoint=True),
          [r'$-1$', r' ', r'$+1$'])
#plt.grid(True)

# Ahora vamos a cambiar los ejes de la figura los ejes

ax = plt.gca() # asigna ax a get current axes
ax.spines['right'].set_color('none')
# Obtenemos acceso a los bordes, colocamos el borde derecho a 'none'
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.legend(loc='upper left')

plt.show()
```
