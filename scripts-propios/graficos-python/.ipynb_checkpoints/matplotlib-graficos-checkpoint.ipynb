```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5.0, 0.1)
x2 = np.arange(0, 5.0, 0.2)
```

```python
plt.figure(1)

plt.subplot(211)
plt.plot(x1, f(x1), 'rs', x2, f(x2), 'k')

plt.subplot(212)
plt.plot(x2, f(x2), 'g--')

plt.show()
```

# Textos en los gráficos

```python
mu = 100
sigma = 20

x = mu+sigma*np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g',
                           alpha=0.6)
# Guardamos los parámetros dentro de las variables de la iz-
#quierda y en los parámetros del histograma se lo agragamso a x
#con 50 divisiones, normalizada, y con una opasidad del 60%

plt.xlabel('Cociente intelectual')
plt.ylabel('Probabilidad')
plt.title(r'Histograma de CI $N(\mu,\sigma)$')
plt.text(30, 0.015, r'$\mu = 100,\; \sigma=20 $')
plt.axis([20,180,0,0.025])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6), dpi = 90)
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.1)
y = np.cos(x)
plt.plot(x,y, 'r')

plt.annotate('Máximo local', xy = (4*np.pi, 1), xytext = (15, 1.5),
            arrowprops = dict(facecolor = 'blue', shrink = 0.5))
plt.grid(True)
plt.ylim(-1.125, 1.125)
plt.show()
```

## Cambios de ejes y escalas
Ahora vamos a trabajar en las diferentes escalas que encontramos en pyplot

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y>1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(1)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale('linear') # Es un parámetro
plt.xscale('linear')
plt.title('Escala lineal')
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale('log')
plt.title('Escala logarítmica')
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y)
plt.yscale('symlog', linthresh = 0.01)
plt.title('Escala log simétrica')
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale('logit')
plt.title('Escala logística')
plt.gca().yaxis.set_minor_formatter(NullFormatter()) # Eliminamos un formato
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom=0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)

plt.show()
```
